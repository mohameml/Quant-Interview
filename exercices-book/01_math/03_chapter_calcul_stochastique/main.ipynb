### LÃ©vy distribution  : 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy, norm

# === Parameters ===
a = 1.0  # hitting level (corresponds to "distance" of Brownian motion)
t = np.linspace(0.001, 5, 500)  # time domain (avoid t=0)

# === Theoretical PDF (LÃ©vy law of hitting time) ===
# f(t) = a / sqrt(2Ï t^3) * exp(-a^2 / (2t))
pdf = (a / np.sqrt(2 * np.pi * t**3)) * np.exp(-a**2 / (2 * t))

# === Using scipy.stats.levy for comparison ===
# Note: Levy(loc=0, scale=1/2a^2) has similar shape (different parameterization)
x = np.linspace(0, 5, 500)
levy_pdf = levy.pdf(x, loc=0, scale=a**2 / 2)

# === Plot ===
plt.figure(figsize=(8, 5))
plt.plot(t, pdf, label=r'Theoretical $f_{\tau_a}(t) = \frac{a}{\sqrt{2\pi t^3}} e^{-a^2/(2t)}$', color='blue', lw=2)
plt.plot(x, levy_pdf, '--', color='red', label='scipy.stats.levy (parametric version)')
plt.title("LÃ©vy Distribution â Hitting Time of Brownian Motion", fontsize=14)
plt.xlabel("t", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
